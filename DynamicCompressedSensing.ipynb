## Experimentation

### Step 1: generated signal

In order to confirm the validity of our implementation, as well as check the algorithm's behavior in a "simple" case, we generate a signal and measurements according to a set of model parameters, that we provide to the algorithm as well.  
Furthermore, we make sure to not use EM to update these parameters during the passes.

```python
from dcsamp.signal_gen import SignalGen
```

```python
gen = SignalGen()
```

```python
x, theta, s = gen.generate_signal()
```

```python
y, A, sig2e = gen.generate_measurements(x)
```

```python
from dcsamp.core import DCS_AMP
from dcsamp.updater import TestUpdater
```

```python
alg = DCS_AMP(updater=TestUpdater())
```

```python
import logging
```

```python
logging.basicConfig(level=logging.DEBUG)
#logging.getLogger('AMP').setLevel(logging.INFO)
```

```python
alg.run(y, A, siggen=gen, sig2e=sig2e)
```
